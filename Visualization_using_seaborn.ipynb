```python
import seaborn as sns
sns.set()
sns.set(style="darkgrid")


import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("https://trello-attachments.s3.amazonaws.com/5f1b8e6a2470af5d1a8d07c3/5f1b8e7cd5a71540a14fa9a2/5749a949a3aa4bb9eda75b14341c1b9d/bigmart_data.csv")
data.head()
```

```python
data = data.dropna(how="any")

data["Visibility_Scaled"] = data["Item_Visibility"] * 100

data.head()
```

#Line Chart
in seaborn lineplot() can be  plot with replot() by setting kind="line"

```python
plt.figure(figsize=(10,8))
sns.lineplot(x="Item_Weight", y="Item_MRP", data=data[:50]);
```

```python
data.Item_Type.value_counts()
```

```python
plt.figure(figsize=(16,8))
sns.barplot(x="Item_Type", y = "Item_MRP", data=data[:50])
```

```python
sns.distplot(data["Item_MRP"]);
```

```python
plt.figure(figsize=(14,8))
sns.boxplot(data=data, orient='vertical');
```

```python
sns.violinplot(data['Item_Outlet_Sales'], orient='vertical', color='green');
```

```python
sns.scatterplot(x='Item_MRP', y ='Item_Outlet_Sales', data=data[:200], color='red' )
```

```python
sns.relplot(x='Item_MRP', y="Item_Outlet_Sales", hue='Item_Type', data=data[:200])
```

```python
sns.relplot(x='Item_MRP', y='Item_Outlet_Sales', data=data[:200], kind='scatter', size='Visibility_Scaled', hue="Visibility_Scaled")
```

```python
sns.relplot(x="Item_Weight", y="Item_Visibility", style="Outlet_Size", col="Outlet_Size", data=data[:200], hue='Outlet_Size');
```

```python
#strip plot
# draw a scatterplot where one variable is categorical

sns.catplot(x='Outlet_Size', y='Item_Outlet_Sales', kind='strip', data=data[:250]);
```

```python
#swarm plot
# similar to strip plot, but points are adjusted, so that don't overlap

sns.catplot(x="Outlet_Size", y='Item_Outlet_Sales', kind='swarm', data=data[:250]);
```
